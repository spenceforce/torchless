Let's start with the function $f(x) = 1$. There is nothing special about this function; it's just simple and a good place to start. If we plot this functions we see a horizontal line at $y=1$.

```python
import matplotlib.pyplot as plt

plt.axhline(y=1, label=r'$f(x) = 1$')
plt.ylim(-0.5, 1.5)
plt.xticks([])
plt.yticks([0, 1])
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Horizontal line')
plt.show()
```

The slope of this line is $0$. This means that $y$ is the same value regardless of $x$. Let's formalize this definition of slope.

> The slope of a horizontal line is $0$.
